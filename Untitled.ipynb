```python
df = pd.read_sql("""
    SELECT
        *,
        json_group_array(
                json_object(
                    'ingredient_id', i.ingredient_id,
                    'name', i.name,
                    'quantity', i.quantity,
                    'unit', i.unit
                )
            ) as w,
        json_object(
            'ingredients', json_group_array(
                json_object(
                    'ingredient_id', i.ingredient_id,
                    'name', i.name,
                    'quantity', i.quantity,
                    'unit', i.unit
                )
            )
        ) AS meal_json
    FROM
        Meals m
    LEFT JOIN
        Ingredients i ON m.meal_id = i.ingredient_id
    LEFT JOIN 
        Directions d ON m.meal_id = d.meal_id
    GROUP BY
        m.meal_id;
    """, conn)
df
```
